```python
import pandas as pd
import requests
import json
from sqlalchemy import create_engine
from config import api_key
```

```python
movies = pd.read_html("https://en.wikipedia.org/wiki/List_of_highest-grossing_films")
df = movies[0]
```

```python
df = df[["Rank", "Title", "Worldwide gross", "Year"]]
df = df.set_index("Rank")
df
```

```python
mylist = df["Title"].tolist()
url = "http://www.omdbapi.com/?t="
director_list = []
rating_list = []
runtime_list = []
imdbrating_list = []
apititle_list = []
```

```python
for title in mylist:
    response = requests.get(url + title + api_key)
    data = response.json()
    try:
        director = data['Director']
        rating = data['Rated']
        runtime = data['Runtime']
        imdbrating = data['imdbRating']
        apititle = data['Title']
        director_list.append(director)
        rating_list.append(rating)
        runtime_list.append(runtime)
        imdbrating_list.append(imdbrating)
        apititle_list.append(apititle)
    except:
        print("That one didn't work.")
        director_list.append("Null")
        rating_list.append("Null")
        runtime_list.append("Null")
        imdbrating_list.append("Null")
        apititle_list.append("Null")
```

```python
df_2 = pd.DataFrame()
```

```python
df_2["title"] = apititle_list
df_2["director"] = director_list
df_2["runtime"] = runtime_list
df_2["rating"] = rating_list
df_2["imdb_rating"] = imdbrating_list
```

```python
df_2.index.name = 'id'
df_2
```

```python
df.index.name="rank"
df.columns=["title", "worldwide_gross", "year"]
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/movie_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
df.to_sql(name='wiki', con=engine, if_exists='append', index=True)
df_2.to_sql(name='omdb', con=engine, if_exists='append', index=True)
```
